```python
word = input("Enter a word: ")

upper_letters = 0
lower_letters = 0

for letter in word:
    if letter.isupper():
        upper_letters += 1
    elif letter.islower():
        lower_letters += 1


print(f"Your word has {upper_letters} capitalized letters and {lower_letters} lower case letters")
```
