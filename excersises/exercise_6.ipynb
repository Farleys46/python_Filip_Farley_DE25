#### 0*

```python
class Convertion:
    def __init__(self, value: int|float):
        self.value = value
        
    def inch_to_cm(self):
        return self.value * 2.54
    
    def foot_to_meters(self):
        return self.value * 0.3048
    
    
test1 = Convertion(6)
print(test1.inch_to_cm())

test1.foot_to_meters()
```
