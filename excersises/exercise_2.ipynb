## 3

```python
number = int(input("Enter your number: "))

if number % 2 == 0:
    print("Your number is even")
else:
    print("Your number is odd")


if number % 5 == 0:
    print("Your number is divisible by 5.")
else:
    print("Your number is not divisible by 5")

if number % 5 == 0 and number % 2 == 0:
    print("Your number is divisible by 5 and an Odd number.")
else:
    print("Your number is not divisible by 5 and an odd number")
```

## 7

```python
import random

user_score = 0

while True:
    rnumber1 = random.randint(0, 10)
    rnumber2 = random.randint(0, 10)

    try:
        user_guess = int(input(f"What is the answer of {rnumber1} * {rnumber2}? "))
    except ValueError:
        print("You must enter a number!")
        continue

    if user_guess == rnumber1 * rnumber2:
        user_score += 1
        print("Good Job! You are correct! You're score is now:",user_score)
    else:
        print("You are incorrect! the correct answer is:",rnumber1 * rnumber2)

    play_again = input("Do you want to play again? (yes/no)").lower()
    if play_again != "yes":
        print(f"Game over! You're final score is:{user_score}")
    break


```
