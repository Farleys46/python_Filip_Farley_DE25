### Translate Testing

```python
# Test shapes:
circle5 = Circle(1, 3, 4.5)
rectangle5 = Rectangle(4, 4, 3, 6)

# Print first positions
print("Before")
print(circle5)
print(rectangle5)

# Use translate method
circle5.translate(-2, -3)
rectangle5.translate(-2, -2)


print("\nAfter translate:")
print(circle5)
print(rectangle5)
```
