### "__repr__" dunder repper method

```python
# __repr__ method

class Student:
    def __init__(self, name, type_of_study, passed, school, age):
        self.name = name
        self.type_of_study = type_of_study
        self.passed = passed
        self.school = school
        self.age = age
        
    def study(self):
        print(f"{self.name} studies {self.type_of_study}") 
    
    def __str__(self):
        return f"{self.name} is studying {self.type_of_study} in school {self.school}"
    
    def __repr__(self):
        # Student (name = "Bobbo"), type_of_study = "economics", school = "Cool School", age = 31, passed = False
        return f"Student (name = '{self.name}', type_of_study = '{self.type_of_study}', school = '{self.school}', age = {self.age}, passed = {self.passed})"


s1 = Student(name = "Anna", type_of_study = "data engineering", passed = True, school = "STI", age = 23)
s2 = Student(name = "Beda", type_of_study = "UX", passed = True, school = "STI", age = 33)



print(s1)
print(s2)

```

```python
class Admission:
    
    # "Dunder init", - initializer - note 2 underscores before and after. 
    # This method will be run when you instantiate a instance from this class 
    def __init__(self, school, name, program = "Cool program"):
        print("Dunder init running")
        
        # Set the attributes to the values from the arguments 
        self.school = school
        self.name = name
        self.program = program

# instantiate the instance person1 that is of type "Admission"
person1 = Admission("STI", "Ceda")
person1.program
person1.name
```
