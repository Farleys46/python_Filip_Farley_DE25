```python
class Student:
    def __init__(student, name, type_of_study, passed, school, age):
        student.name = name
        student.type_of_study = type_of_study
        student.passed = passed
        student.school = school
        student.age = age
        
s1 = Student(name = "Anna", type_of_study = "data engineering", passed = True, school = "STI", age = 23)
s2 = Student(name = "Beda", type_of_study = "UX", passed = True, school = "STI", age = 33)

print(s1.school,"\n",
    s2.name,"\n",
    s1.name)


```

```python
class Admission:
    
    # "Dunder init", - initializer - note 2 underscores before and after. 
    # This method will be run when you instantiate a instance from this class 
    def __init__(self, school, name, program = "Cool program"):
        print("Dunder init running")
        
        # Set the attributes to the values from the arguments 
        self.school = school
        self.name = name
        self.program = program

# instantiate the instance person1 that is of type "Admission"
person1 = Admission("STI", "Ceda")
person1.program
person1.name
```

```python
person2 = Admission(name = "Johnny", school = "Cool school", program = "Java")

person2
```
