## Matplotlib
- focus on the distinction between two ways of writing matplotlib codes
- focus on the components of a matplotlib figure 

```python
import matplotlib.pyplot as plt
import numpy as np

# data type - arrays 
```

##### The first approach - functional/matlab style

```python
# generate data
x = np.linspace(-5, 5, 100)
x
```

```python
# generate data f

f = np.sin(x)
f
```

```python
plt.figure()
plt.plot(x, f)
plt.title("sin(x)")
plt.xlabel("x")
plt.ylabel("y")
```

#### The second approach - OOP 

```python
fig = plt.figure() # create a figure   
ax = plt.axes()  # in the figure, create an axe
ax.plot(x, f) # on the axe, create a plot
ax.set(xlim=(-10, 10), ylim=(-5, 5), xlabel="x", ylabel="y") # continue modify the axe object

```
