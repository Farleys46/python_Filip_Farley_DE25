## Reading a file

```python
#### Open the file with 'with' statement

f = open("test.txt", "r")
print(f)

### Save the contents of the file in a Variable
text = f.read()
print(text)

f.close()
```

```python
### Open with "with" statement

with open("test.txt", "r") as f:
    text = f.read()
    print(text)
```
