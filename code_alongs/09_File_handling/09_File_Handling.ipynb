## File Handling

- read the file
- clean contents
- write the clean contents to a new text file

## Reading a file

```python
#### Open the file with 'with' statement

f = open("test.txt", "r")
print(f)

### Save the contents of the file in a Variable
text = f.read()
print(text)

f.close()
```

```python
### Open with "with" statement

with open("test.txt", "r") as f:
    text = f.read()
    print(text)
```

```python
import re

clean_quotes = []
i = 1 

with open("quotes.txt", "r") as f_read, open("quotes_cleaned.txt", "w") as f_write:
    for quotes in f_read:
        quotes = quotes.strip(" \n")
        quotes = re.sub(" +", " ", quotes)
        print(quotes)
        
        if quotes != "":
            f_write.write(f"{i}. {quotes}\n")
            i += 1


```
